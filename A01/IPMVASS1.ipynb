```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('emma.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None

transform1 = np.linspace(0,50,50)
transform2 = np.linspace(100,255,100)
transform3 = np.linspace(150,255,106)
transform = np.concatenate((transform1,transform2,transform3), axis = 0).astype(np.uint8)

transformed = cv.LUT(img,transform)

assert len(transform)==256
im = cv.cvtColor(img, cv.COLOR_BGR2RGB)
im2 = cv.cvtColor(transformed, cv.COLOR_BGR2RGB)
fig, ax = plt.subplots(1,3, figsize = (20,20))
ax[0].imshow(im)
ax[1].imshow(im2)
ax[0].set_title('Before')
ax[1].set_title('After')
ax[2].plot(transform)
ax[2].set_aspect('equal')
plt.show()
```

2. Applying an intensity transform for a brain proton density slice

a) white matter

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('BrainProtonDensitySlice9.png', cv.IMREAD_GRAYSCALE)
Converted = cv.cvtColor(img, cv.COLOR_BGR2RGB)
fig, ax = plt.subplots(1,3,figsize = (20,20))
ax[0].imshow(Converted)

transform1 = np.linspace(0,25,210)
transform2 = np.linspace(25,240,20)
transform3 = np.linspace(240,255,26)
transform = np.concatenate((transform1,transform2,transform3), axis = 0).astype(np.uint8)

transformed = cv.LUT(img,transform)

CvtTfd = cv.cvtColor(transformed, cv.COLOR_BGR2RGB)
ax[2].imshow(CvtTfd)
ax[1].plot(transform)
ax[1].set_aspect('equal')
ax[0].set_title('Original')
ax[1].set_title('Transformation applied')
ax[2].set_title('White matter')
plt.show()
```
